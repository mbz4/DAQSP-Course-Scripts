```python
import matplotlib.pyplot as plt
import numpy as np

data_file = input("Please, input data.txt file you want to plot: ")

f = open(data_file, "r")

data = f.read()

# this function has a loop that iterates thru the entries in data and converts each to float, saves to an array
def txt_to_float(data_input):
    data_sp = data_input.split()
    fdata = []
    for entry in data_sp:
        fdata.append(float(entry))
    return fdata

# this function takes in filename and returns output plot name without .txt appendix
def output_plot_filename(file_name, sample_nm):
    name_temp = file_name.rsplit('.',1)[0]
    samples = str(sample_nm)
    name_out = "plot  " + name_temp + "  # of samples= " + samples + ".png"
    return name_out

# this function counts the number of lines in the text file
def sample_nm(from_data):
    counter = 0
    CoList = from_data.split("\n")
    for i in CoList:
        if i:
            counter += 1

    return counter

samples = sample_nm(data)

# Data for plotting
t = np.arange(0, samples)
s = txt_to_float(data)

#plot setup
fig, ax = plt.subplots()
ax.plot(t, s)

# plt.xticks(t) # in case you want to read voltage for each sample individually

ax.set(xlabel='# of Samples', ylabel='voltage (V)',
       title=data_file)
ax.grid()


fig = plt.gcf()
fig.set_size_inches(22, 14, forward=True)

plt.show()

plot_name = output_plot_filename(data_file, samples)

fig.savefig(plot_name, dpi=300)
print("Saved plot as: " + plot_name)

f.close()
```
